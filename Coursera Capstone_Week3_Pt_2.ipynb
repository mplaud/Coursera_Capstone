```python
for mydict in Canada:
    Postalcode = mydict["Postcode"]
    Borough_name = mydict["Borough"]
    Neighbourhood_name = mydict["Neighbourhood"] 
    
    
    postalcanada = postalcanada.append({'Postcode': Postalcode,
                                          'Borough': Borough_name,
                                          'Neighbourhood': Neighbourhood_name},
                                           ignore_index=True)
postalcanada.head()
```

### We remove the "Not assigned" Borough row and update the Queen's Park Neighbourhood value

```python
postcan = postalcanada[postalcanada.Borough != 'Not assigned']
postcan["Neighbourhood"][8] = "Queen's Park"
postcan.head()
```

### We combine Boroughs with the same postal code and join them with comma's

```python
df = postcan.groupby('Postcode', sort=False, as_index=False).agg({'Borough':'first','Neighbourhood':','.join})
df.reset_index()
df.head(20)
```

```python
df.insert(3,'Latitude',None, allow_duplicates = False)
df.insert(4,'Longitude',None, allow_duplicates = False)

df.head()


```
